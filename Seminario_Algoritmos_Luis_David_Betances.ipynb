- <span style='font-family:Georgia'><font color='Orange'>**(*)**</font> <span style='font-family:Georgia'><font color='Black'>**¿Cual es la función objetivo?**</font>
---
<span style='font-family:Georgia'><font color='Red'><u>Respuesta:</u></font>

- Sea $S$ la variable aleatoria que define las sesiones de grabación, con índice $s \in S$.

- Sea $n_s$ el número máximo de tomas permitidas en la sesión de grabación $s$, con $n_s \in \mathbb{N}$.

- Sea $A$ la variable aleatoria que define los actores, con índice $a \in A$.

- Sea $T$ la variable aleatoria que define las tomas, con índice $t \in T$.

<br>

**Variables de decisión:**

- Para todo $a \in A$ y $t \in T$, $δ_{at} = 
  \begin{cases} 
      1, \mbox{ si el actor } a \mbox{ interviene en la toma } t  \\
      0, \mbox{ en otro caso } 
   \end{cases}$

- Para todo $a \in A$ y $s \in S$, $x_{as} = 
  \begin{cases} 
      1, \mbox{ si el actor } a \mbox{ esta en la sesion } s  \\
      0, \mbox{ en otro caso } 
   \end{cases}$

- Para todo $t \in T$ y $s \in S$, $z_{ts} = 
  \begin{cases} 
      1, \mbox{ si la toma } t \mbox{ esta en la sesion } s  \\
      0, \mbox{ en otro caso } 
   \end{cases}$

- Para todo $a \in A$, $t \in T$ y $s \in S$, $y_{ats} = 
  \begin{cases} 
      1, \mbox{ si el actor } a \mbox{ interviene en la toma } t \mbox{ en la sesion } s \\
      0, \mbox{ en otro caso } 
   \end{cases}$

<br>

**RESTRICCIONES:**

- $\sum_{s \in S}y_{ats} = δ_{at}  \hspace{0.7cm} \forall a \in A, \forall t \in T$ 

- $\sum_{t \in T}z_{ts} \leq n_s  \hspace{1cm} \forall s\in S$ 

- $y_{ats} \leq x_{as}  \hspace{1.7cm} \forall a\in A,\forall t\in T,\forall s\in S$ 

- $y_{ats} \leq z_{ts}  \hspace{1.8cm} \forall a\in A,\forall t\in T,\forall s\in S$ 




<br>

**FUNCION OBJETIVO**:

<font color="red">$zmin = \sum_{a \in A} \sum_{s \in S} x_{as}$</font>

````python

# Función auxiliar para usar como decorador y calcular tiempo de ejecución -----------------------------------------------

def calcular_tiempo(f):
  """Decorador para calcular tiempo de ejecución"""
  def new_function(*args):
    from time import time
    start = time()
    f(*args)
    print(chr(27)+'[0;32m' + '\nTiempo de ejecución: ' + str(time()-start) + '\n')
  return new_function
# ------------------------------------------------------------------------------------------------------------------------
````

````python

# importar librerías -----------------------------------------------------------------------------------------------------
import pandas as pd
from itertools import combinations
# ------------------------------------------------------------------------------------------------------------------------
````

````python
# ------------------------------------------------------------------------------------------------------------------------
# importar datos con las 30 tomas y los 10 actores
path = 'https://raw.githubusercontent.com/LDBM42/03MAIR_Algoritmo_Optimizacion_PUBLICO/master/datos_problema_doblaje.csv'
df_tomas = pd.read_table(path, sep=',', usecols=list(range(1,11)), header=None)

df_tomas.drop(index=[0,1,32,33], inplace=True) # borrar filas no deseadas
df_tomas.index = list(range(1,31)) # nombrar indices del 1 al 30 (Tomas)
df_tomas = df_tomas.astype(int) # convertir valores a tipo entero

cantidad_de_tomas = 12
df_tomas = df_tomas[0:cantidad_de_tomas] # LIMITAR CANTIDAD DE TOMAS
display(df_tomas) # mostrar tabla
# ------------------------------------------------------------------------------------------------------------------------
````

````python

# FUNCION PRINCIPAL ------------------------------------------------------------------------------------------------------
@calcular_tiempo  
def fuerza_bruta_doblaje(df_tomas, actores, maximas_tomas_por_dias):
  """Organiza las sesiones de doblaje por fuerza bruta.
     La solución se mostrará de la siguiente manera:
     (6, 7, 9) 4             <- (6, 7, 9) son las tomas por día con menor coste.  
     (1, 2, 3) 6             <-  6 es la cantidad de actores en todas las tomas 
     El coste total es: 10   <-  10 es la suma de todos los actores por día (coste total)
  """
  from itertools import combinations
  
  tomas = df_tomas.copy() # copiar tomas
  total_actores = 0 # para calcular el total de actores
  mejor_valor = 10e10 
  mejor_solucion = ()
  coste_total = 0 # para calcular el coste total

  # iteramos sobre las tomas entre el maximo de dias, para optener una solución por día
  for _ in range(len(tomas)//maximas_tomas_por_dias): 
    # iteramos sobre todas las posibles permutaciones
    for tomas_posibles in combinations(list(tomas.index), maximas_tomas_por_dias): 
      # iteramos sobre cada actor para calcular el total de actores por tomas
      for actor in range(1, actores+1):
        # sumamos la cantidad de veces que aparece 'x' actor en todas las tomas
        actores_por_tomas = tomas.loc[list(tomas_posibles)][actor].sum()
        if actores_por_tomas > 0: # si 'x' actor aparece por lo menos 1 vez lo contamos 
          total_actores = total_actores+1 # ir acumulando total de actores en todas las tomas

      # calcular mejor valor y solución por día
      if total_actores < mejor_valor:
        mejor_valor = total_actores
        mejor_solucion = tomas_posibles
      
      total_actores = 0 # reiniciar contador

    coste_total = coste_total + mejor_valor # ir acumulando coste total
      
    print(mejor_solucion, mejor_valor)
    mejor_valor = 10e10 # reiniciar mejor valor
    tomas.drop(list(mejor_solucion), inplace=True) # borrar tomas con mejor solución por dia
  print('El coste total es:', coste_total)
# ------------------------------------------------------------------------------------------------------------------------
````

````python

# ------------------------------------------------------------------------------------------------------------------------

actores = 10
maximas_tomas_por_dias = 3
print(chr(27)+"[1;31m" + '\nSESION DOBLAJE (POR FUERZA BRUTA) | ' + str(cantidad_de_tomas) + 
      ' tomas totales y ' + str(maximas_tomas_por_dias) + ' tomas por día |:\n' + chr(27)+"[0m")
fuerza_bruta_doblaje(df_tomas, actores, maximas_tomas_por_dias) # llamado a función

# ------------------------------------------------------------------------------------------------------------------------
````

- <span style='font-family:Georgia'><font color='Black'>**Calcula la complejidad del algoritmo por fuerza bruta**</font>
---
<span style='font-family:Georgia'><font color='Red'><u>Respuesta:</u></font>

Al calcular la complejidad (Big-O), las asignaciones, las operaciones aritméticas básicas y cualquier otra operación con complejidad O(1), carece de importancia si existe alguna otra operación con mayor complejidad, por lo que me centraré en la parte del algoritmo con complejidad mayor a esta.

<img src=https://raw.githubusercontent.com/LDBM42/03MAIR_Algoritmo_Optimizacion_PUBLICO/master/SesionDoblaje_FuerzaBruta_.png width="800">


**n** = tomas;

**x** = maximas_tomas_por_dia; 

**a** = actores

<font color='Red'>**(1)**</font>
```python
for _ in range(len(tomas)//maximas_tomas_por_dias): 
```
Esto es igual a O(n/x), pero ya que en Big-O generalmente nos quedamos con el peor resultado posible, podríamos trabajar sobre el caso de que x=1, pero como en el paso (2) la complejidad mayor sería si x=n/2 y es una complejidad con mayor crecimiento que esta, nos quedaría O(n/x) = O(n/(n/2)):

**complejidad**: $O\biggl( \frac{n}{\frac{n}{2}} \biggl)$


<br/>

<font color='Red'>**(2)**</font>
```python
for tomas_posibles in combinations(list(tomas.index), maximas_tomas_por_dias): 
```
Combinaciones ($C_{n,x}$) = $\frac{n!}{x!(n-x)!}$    

el peor caso sería si $x = \frac{n}{2} $ por lo que: $\frac{n!}{x!(n-x)!} = \frac{n!}{\frac{n}{2}!(n-\frac{n}{2})!} = \frac{n!}{\frac{n}{2}!\frac{n}{2}!}$

**complejidad**: $O\biggl( \frac{n!}{\frac{n}{2}!\frac{n}{2}!} \biggl)$

<br/>

<font color='Red'>**(3)**</font>
```python
for actor in range(1, actores+1):
```
La cantidad de actores en este caso es 10, pero podría variar por lo que la complejidad no es constante, sino O(a).

**complejidad**: O(a)

<br/>

<font color='Red'>**(4)**</font>
```python
actores_por_tomas = tomas.loc[list(tomas_posibles)][actor].sum()
```
La operación sum() de python tiene una complejidad O(n).

**complejidad**: O(n)

````python

# ------------------------------------------------------------------------------------------------------------------------
# Importar librerías 
import numpy as np
import pandas as pd
from time import time
from itertools import combinations
# ------------------------------------------------------------------------------------------------------------------------
````

````python

# ------------------------------------------------------------------------------------------------------------------------
# importar datos con las 30 tomas y los 10 actores
path = 'https://raw.githubusercontent.com/LDBM42/03MAIR_Algoritmo_Optimizacion_PUBLICO/master/datos_problema_doblaje.csv'
df_tomas = pd.read_table(path, sep=',', usecols=list(range(1,11)), header=None)

df_tomas.drop(index=[0,1,32,33], inplace=True) # borrar filas no deseadas
df_tomas.index = list(range(1,31)) # nombrar indices del 1 al 30 (Tomas)
df_tomas = df_tomas.astype(int) # convertir valores a tipo entero

# Esto en caso de querer limitar la cantidad de tomas --------------------------------------------------------------------
# tomas = 18 # ej.: 18 tomas
# df_tomas = df_tomas[0:tomas] # Esto en caso de querer limitar la cantidad de tomas
display(df_tomas) # mostrar tabla
# ------------------------------------------------------------------------------------------------------------------------
````

````python

# ------------------------------------------------------------------------------------------------------------------------
def menor_gasto(tomas:'numpy_array', tomas_actores:'DataFrame', solucion_final:'numpy_array') -> 'numpy_array':
    """Evalúa si el gasto de la solución parcial es el menor de las soluciones anteriores"""
    
    # array de soluciones (0 o 1) dependiendo de si la toma está elegida o no
    solucion = np.array([1 if(i in tomas) else 0 for i in range(1, len(solucion_final))])
    actores = tomas_actores.loc[tomas] # seleccionar solo los actores de las tomas elegidas

    gastos = 0
    # Calcular los gastos por día
    for i in range(1,len(actores.columns)+1):
        if sum(actores[i]) != 0:
            gastos = gastos + 1

    # actualizar solución final si la solución parcial tiene menos gastos
    if gastos < solucion_final[-1]:
        solucion_final[:] = np.concatenate((solucion, [gastos]))
        return tomas # retornar la toma
    
    return None
# ------------------------------------------------------------------------------------------------------------------------

# ------------------------------------------------------------------------------------------------------------------------
def sesiones_doblaje_por_dia(tomas_maximas:int, tomas_actores:'DataFrame', 
                             solucion_final:'numpy_array') -> 'numpy_array':
    """obtiene la mejor sesión de doblaje por día"""
    tomas_menor_gasto = [] # tomas con menor gasto
    # iterar sobre todas las combinaciones sin repetición
    for tomas in combinations(tomas_actores.index, tomas_maximas): 
        tomas = np.asarray(tomas) # convertir tupla a array
        
        _ = menor_gasto(tomas, tomas_actores, solucion_final) # actualizar solución final
        if(_ is not None): tomas_menor_gasto = _ # almacenar toma si es la menor
    return tomas_menor_gasto
# ------------------------------------------------------------------------------------------------------------------------

# ------------------------------------------------------------------------------------------------------------------------        
def imprimir_organizar_sesiones_doblaje(tomas_maximas:int, total_tomas:int, solucion_total:'numpy_array'):
    """Imprime la organización de las sessiones de doblaje, mostrando las tomas por días y sus gastos"""
    # crear nombre de las columnas, header e idices del dataframe    
    cols = np.arange(1,total_tomas+1)
    cols = np.append(cols, ['Gastos ↓'])
    header = pd.MultiIndex.from_product([['Tomas ↓'], cols],
                                        names=['Días ↓',' '])
    cantidad_dias = total_tomas//tomas_maximas # cantidad de días para grabar
    index = np.arange(1,cantidad_dias+1) # los indices seran los días
    
    # crear dataframe
    df_solucion_total = pd.DataFrame(solucion_total, columns=header, index=index) 
    
    # imprimir gasto total
    total = np.full(total_tomas, ' ')
    total[0] = '→'
    gasto_total = df_solucion_total[('Tomas ↓', 'Gastos ↓')].sum()
    total = np.append(total, gasto_total) # agregar dato total al array
    df_solucion_total.loc['Total'] = total # agregar fila con gasto total al DataFrame
    
    # Mostrar dataframe con resultado final
    pd.set_option('display.max_columns',None) # mostrar todas las columnas (evitar el ...)
    display(df_solucion_total)
# ------------------------------------------------------------------------------------------------------------------------    
    
# ------------------------------------------------------------------------------------------------------------------------    
# la solución final tendrá las tomas por día y el coste en la última columna
def organizar_sesiones_doblaje(tomas_maximas:int, total_tomas:int, tomas_actores:'DataFrame'):
    """obtiene la organización de las sessiones de doblaje con menor gasto"""
    
    # inicializar solución parcial y final
    solucion=np.full(total_tomas,-1) # arreglo con tamaño 'total_tomas' lleno de '-1'
    solucion_por_dia = np.concatenate((np.full(total_tomas,-1), [np.inf])) # [1,0,1,0,0,...,1,15] donde 15 es el gasto por día.
    
    solucion_total  = solucion_por_dia.copy()
    tomas_actores_c = tomas_actores.copy() # hacer una copia de las tomas y actores
    
    for _ in range(total_tomas//tomas_maximas):
        # calcular solución por día y optener las tomas
        solucion_por_dia[-1] = np.inf # reiniciar gastos
        tomas = sesiones_doblaje_por_dia(tomas_maximas, tomas_actores_c, solucion_por_dia) # llamado a Función
        solucion_total = np.vstack((solucion_total, solucion_por_dia))
        # borrar tomas elegidas
        tomas_actores_c = tomas_actores_c.drop(tomas)     
    
    solucion_total = solucion_total[1:].astype(int) # convertimos los valores a enteros    
    imprimir_organizar_sesiones_doblaje(tomas_maximas, total_tomas, solucion_total)
# ------------------------------------------------------------------------------------------------------------------------
````

````python

# ------------------------------------------------------------------------------------------------------------------------
def organizar_sesiones_doblaje_final(tomas_maximas:int, total_tomas:int, titulo:str, tomas_actores:'DataFrame'):
  # Imprimir título ------------------------------------------------------------------------------------------------------
  print('\n\n' + chr(27)+'[1;31m' + titulo + (chr(0x2015)*(116-len(titulo))) + '\n')

  # ----------------------------------------------------------------------------------------------------------------------
  start = time() # iniciar cronometro

  # llamado a función
  organizar_sesiones_doblaje(tomas_maximas, total_tomas, tomas_actores)

  # imprimir tiempo de ejecución
  t_e = 'Tiempo de ejecución: ' + str(time()-start) + ' '
  print('\n' + chr(27)+'[0;32m' + t_e + chr(0x2015)*(116-len(t_e)) + '\n')   
  # ----------------------------------------------------------------------------------------------------------------------


# ------------------------------------------------------------------------------------------------------------------------
"""------NOTA: EL ALGORITMO CON LAS 30 TOMAS TOTALES Y 6 TOMAS POR DIA, DURA UNOS 10 MINUTOS,
               EL TIEMPO DE EJECUCION SE REDUCE CONSIDERABLEMENTE SI SE ELIGE SOLO < 3 > TOMAS POR DIA-------"""
maximas_tomas_por_dias = 6
total_tomas = len(df_tomas)
titulo = 'ORGANIZAR SESIONES DE DOBLAJE | ' + str(total_tomas) + ' tomas totales y ' + str(maximas_tomas_por_dias) + ' tomas por día | '

organizar_sesiones_doblaje_final(maximas_tomas_por_dias, total_tomas, titulo, df_tomas)
# ------------------------------------------------------------------------------------------------------------------------
````

Combinaciones ($C_{n,x}$) = $\frac{n!}{x!(n-x)!}$
